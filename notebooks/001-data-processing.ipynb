```python
# Check null for each column
null_counts = df.isnull().sum()
total_rows = df.shape[0]

# Drop columns with more than 30% null values
columns_to_drop = null_counts[null_counts > 0.3 * total_rows].index
df = df.drop(columns=columns_to_drop)

df.head()
```

```python
null_counts = df.isnull().sum().sort_values(ascending=False)

print(null_counts)
```

```python
df = df.drop(columns=['id', 'schedule_month', 'schedule_week', 'schedule_year', 'schedule_hour', 'schedule_weekday', 'schedule_date', 'schedule_time'])
```

```python
df.head()
```

```python
df = df.drop(columns=['status_process', 'status_partner', 'schedule_id', 'queue', 'forward', 'sync_status', 'district', 'updated_time', 'service', 'bc_id', 'place_id', 'pay_type'])
```

```python
null_counts = df.isnull().sum().sort_values(ascending=False)

print(null_counts)
```

```python
df = df.drop(columns=['created_hour', 'created_year', 'created_month', 'created_date', 'created_week', 'created_weekday'])
```

```python
df = df.drop(columns=['partner_update_lasttime', 'partner_update_status'])
```

```python
df.head()
```

```python
df = df.drop(columns=['cancel_by', 'book_method', 'scheduled_time', 'pin', 'referrer', 'device', 'is_delete', 'auto_confirm', 'view_guide_last', 'book_for', 'status_contact'])
```

```python
df.head()
```

```python
df['created_time'] = pd.to_datetime(df['created_time'], unit='s', errors='coerce')
df['created_time'] = df['created_time'].dt.year
df['birthtime'] = pd.to_datetime(df['birthtime'], errors='coerce')
df['birthtime'] = df['birthtime'].dt.year
df['age'] = df['created_time'] - df['birthtime']
df.head()
```

```python
df = df.drop(columns=['created_time', 'birthtime'])
df.head()
```

```python
null_counts = df.isnull().sum().sort_values(ascending=False)

print(null_counts)
```

```python
df['reason_combind'] = df['reason'].combine_first(df['reason_other'])
```

```python
df
```

```python
df_cleaned = df.dropna(subset=['reason_combind'])
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned.drop(columns=['reason', 'reason_other'])
```
