```python
specialist_df = pd.read_csv('../data/tbl_clinic_specialist.csv')
```

```python
data_dict
```

```python
import json
json_path = '../data/specialist.json'
with open(json_path, 'w', encoding='utf-8') as f:
    json.dump(data_dict, f, ensure_ascii=False, indent=4)
```
