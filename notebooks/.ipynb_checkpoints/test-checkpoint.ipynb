```python
import os
import pandas as pd
import threading

# Get all CSV files in data_version1 directory
csv_files = []
excluded_files = []

for file in os.listdir('../data/data_version1/'):
    if file.endswith('.csv') and file.startswith('reason_specialist'):
        if not any(excluded in file for excluded in excluded_files):
            csv_files.append(file)

print(f"Found {len(csv_files)} CSV files to process: {csv_files}")

# Process each file in a separate thread
def process_all_files():
    threads = []
    
    # Create an empty list to store all dataframes
    all_dfs = []
    
    for i, file in enumerate(csv_files):
        # Read the file
        file_path = f'../data/data_version1/{file}'
        df = pd.read_csv(file_path, index_col=0)
        df = df.reset_index(drop=True)
        
        # Add the dataframe to the list
        all_dfs.append(df)
    
    # Merge all dataframes into one
    merged_df = pd.concat(all_dfs, ignore_index=True)
    
    return merged_df


# Start processing
df = process_all_files()
print("All files merged successfully")
```

```python
import json

# Define the output file path
output_file = '../data/data_version3/dataset.jsonl'

# Process DataFrame and create JSONL entries
with open(output_file, 'w', encoding='utf-8') as f:
    for idx, row in df.iterrows():
        # Format the user prompt with data from the row
        formatted_user_prompt = user_prompt.format(
            reason=row.get('reason_combind', ''),
            specialist=row.get('specialist_name', '')
        )
        
        # Create a JSON object for each row
        json_entry = {
            "custom_id": f"task-{idx}",
            "method": "POST",
            "url": "/chat/completions",
            "body": {
                "model": "gpt-4o-mini",
                "messages": [
                    {
                        "role": "system",
                        "content": system_prompt
                    },
                    {
                        "role": "user",
                        "content": formatted_user_prompt
                    }
                ]
            }
        }
        
        # Write the JSON object to the file
        f.write(json.dumps(json_entry) + '\n')

print(f"JSONL file created successfully at {output_file}")
```

```python
system_prompt = """
You are a medical text processor that extracts and standardizes Vietnamese symptoms and diseases from examination reasons.

# Input Format
{
  "records": [
    {
      "reason_combined": "Patient's stated reason for examination",
      "specialist_name": "Medical specialty being consulted"
    },
    {...}
  ]
}

# Core Rules
1. Extract only ACTUAL medical conditions that are CURRENTLY present
2. Correct spelling errors (e.g., "bung" → "bụng", "xuyễn" → "suyễn")
3. Format as comma-separated list IN VIETNAMESE ONLY: "symptom1, symptom2, disease1, disease2"
4. Return null when conditions don't warrant extraction
5. ALWAYS output in Vietnamese language, never translate to English or other languages

# When to Return null
- Symptoms/diseases don't match the specialist field
- No current symptoms mentioned (except for "nội khoa" specialty)
- Past medical events without current symptoms ("đã mổ...", "tiền sử...")
- Administrative requests, routine follow-ups, or record requests
- Test results without associated symptoms

# Special Cases
- For "nội khoa" (internal medicine) only: "khám tổng quát" (general checkup) is valid and returned as is
- Never change the medical meaning - correct spelling but preserve the condition

# Response Format
Provide ONLY a valid JSON object with this structure:
{
  "results": [
    {"result": "extracted text or null"},
    {"result": "extracted text or null"},
    ...
  ]
}

# Example Input/Output
Input:
{
  "records": [
    {"reason_combined": "Đau đầu, chóng mặt, buồn nôn", "specialist_name": "Thần kinh"},
    {"reason_combined": "Xuất huyết dạ dày, đau bung", "specialist_name": "Tiêu hóa"},
    {"reason_combined": "Khám tổng quát", "specialist_name": "Nội khoa"},
    {"reason_combined": "Đã mổ ruột thừa 2 năm trước", "specialist_name": "Ngoại khoa"}
  ]
}

Output:
{
  "results": [
    {"result": "đau đầu, chóng mặt, buồn nôn"},
    {"result": "xuất huyết dạ dày, đau bụng"},
    {"result": "khám tổng quát"},
    {"result": null}
  ]
}
"""
```

```python
import json
from tqdm.notebook import tqdm
from openai import OpenAI

model="Qwen/Qwen2.5-14B-Instruct-AWQ"
client = OpenAI(
    api_key="UNKOWN",
    base_url="https://55df-104-196-51-130.ngrok-free.app/v1"
)

def batch_process_records(dataframe, batch_size=100):
    """Process the dataframe in batches and return combined results."""
    total_records = len(dataframe)
    all_results = []
    
    for start_idx in tqdm(range(0, total_records, batch_size), desc="Processing data..."):
        end_idx = min(start_idx + batch_size, total_records)
        print(f"Processing batch {start_idx//batch_size + 1}, records {start_idx} to {end_idx-1}")
        
        # Extract batch records
        batch_df = dataframe.iloc[start_idx:end_idx].copy()
        
        # Create input structure for the batch
        batch_records = []
        for _, row in batch_df.iterrows():
            batch_records.append({
                "reason_combined": row.get('reason_combind', ''),
                "specialist_name": row.get('specialist_name', '')
            })
        
        input_json = {"records": batch_records}
        print()
        
        # Process batch with Azure OpenAI
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": "Input: " + json.dumps(input_json)}
            ],
            temperature=0.0,
            frequency_penalty=0.0,
            presence_penalty=0.0,
            max_tokens=4000,
            response_format={"type": "json_object"},
        )
        
        # Parse the response
        try:
            results_text = response.choices[0].message.content
            # Handle both JSON and string responses
            try:
                results_data = json.loads(results_text)
                print(results_data)
                if "results" in results_data:
                    all_results.extend(results_data["results"])
                else:
                    print(f"Warning: Unexpected response format in batch {start_idx//batch_size + 1}")
                    placeholder_results = [{"result": None} for _ in range(len(batch_df))]
                    all_results.extend(placeholder_results)
            except json.JSONDecodeError:
                # If not JSON, treat each line as a separate result
                print(f"Warning: Response is not valid JSON in batch {start_idx//batch_size + 1}")
                lines = results_text.strip().split('\n')
                batch_results = []
                for line in lines[:len(batch_df)]:  # Limit to batch size
                    batch_results.append({"result": line if line.strip() else None})
                # Pad with None if needed
                while len(batch_results) < len(batch_df):
                    batch_results.append({"result": None})
                all_results.extend(batch_results)
        except Exception as e:
            print(f"Error processing response in batch {start_idx//batch_size + 1}: {str(e)}")
            placeholder_results = [{"result": None} for _ in range(len(batch_df))]
            all_results.extend(placeholder_results)
    
    return all_results

# Process the entire dataset in batches
# Test with a smaller subset of 100 records
batch_size = 10
test = df.head(100)
results = batch_process_records(test, batch_size)

# Add results to the test dataframe - handle different possible result formats
test['processed_symptoms'] = [item.get('result') if isinstance(item, dict) else item for item in results]

# Save the test dataframe
test.to_csv('../data/data_version3/dataset_processed.csv', index=False)

print(f"Test processing complete. Processed {len(test)} records.")

# Examine a few results
print("\nSample results:")
for i in range(min(5, len(test))):
    print(f"Record {i+1}:")
    print(f"Reason: {test.iloc[i]['reason_combind']}")
    print(f"Specialist: {test.iloc[i]['specialist_name']}")
    print()
```
